```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Constructing the dataframe
data = {
    'comparison': ['MASPpepK VS adjuvant', 'NGP24h VS adjuvant', 'MASPpepK + NGP24h VS adjuvant'],
    'increased': [2, 2, 0],
    'decreased': [41, 50, 1]
}
df = pd.DataFrame(data)

# Melting the dataframe for seaborn
df_melted = pd.melt(df, id_vars='comparison', value_vars=['increased', 'decreased'], var_name='change', value_name='count')


plt.figure(figsize=(10, 3))
plot_1 = sns.barplot(y='comparison', x='count', hue='change', data=df_melted)
plt.title('', fontsize = 13)
plt.ylabel('')
plt.xlabel('Number of CARs affected by different vaccine groups', fontsize = 13)
plt.xticks()
plt.yticks(fontsize = 13)
plot_1.spines['top'].set_visible(False)
plot_1.spines['right'].set_visible(False)
plt.legend(title='')
plt.tight_layout()
plt.savefig("barplot_CAR.png", dpi=300, bbox_inches='tight')
plt.show()
```
