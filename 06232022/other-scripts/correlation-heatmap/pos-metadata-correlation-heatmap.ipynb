```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data_a = pd.read_csv("06232022-metadata.csv")
data_a = data_a.set_index('sample_ID')
sns.set(rc = {'figure.figsize':(10,8)})
sns.set(font_scale=1.7)
g = sns.heatmap(data_a.corr(method='spearman'), annot=True, annot_kws={"fontsize":10}, vmin=-1, vmax=1, center= 0, cmap = 'viridis')

plt.savefig("pos-metadata.pdf", dpi = 300, bbox_inches = 'tight')
plt.savefig("pos-metadata.png", dpi = 300, bbox_inches = 'tight')
```
