```r
a <- read.csv("06232022-pos-M-F-filtered.csv")
```

```r
head(a)
dim(a)
```

```r
features_pos<-droplevels(subset(a, a$Group_Short==c("MASPpepK")|a$Group_Short==c("adjuvant")))
head(features_pos)
dim(features_pos)
```

```r
noMeta<-features_pos[,26:2082]
head(noMeta)
not_zero<-which(colSums(noMeta) != 0)
head(not_zero)
nonzero <- noMeta[, not_zero]
nonzero
nonzero_wMeta<-cbind(features_pos[,1:25], nonzero)
head(nonzero_wMeta)
dim(features_pos)
dim(noMeta)
dim(nonzero)
dim(nonzero_wMeta)
```

```r
pval<-c()
for (i in 26:ncol(nonzero_wMeta)) 
pval<-c(pval,wilcox.test(nonzero_wMeta[,i]~nonzero_wMeta$Group_Short)$p.value)
head(pval)
pAdj <- p.adjust(pval, method = c("fdr"), n=length(pval)) 
head(pAdj) 
```

```r
length(pval)
```

```r
length(pAdj)
```

```r
avg<- apply(nonzero_wMeta[,26:ncol(nonzero_wMeta)], 2, function(x) tapply(x, nonzero_wMeta$Group_Short, function(x) mean(x))) 
dim(avg) 
avg[,1:3]
fch <- apply(avg, 2, function(x) x[1]/x[2])
head(fch)
is(fch)
```

```r
fch
```

```r
df <- data.frame(fch,pval, pAdj)
write.csv(df, "pos-filtered-all-wilcox-MA-adj.csv")
```
