```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time

# change the file name to yours, make sure the cluster index column name is "cluster_index"
data_feature = pd.read_csv("test-feature-data.csv")
feature_list = data_feature.cluster_index.to_list()
print(feature_list)
driver = webdriver.Chrome()
# change GNPS link to yours
driver.get("https://gnps.ucsd.edu/ProteoSAFe/status.jsp?task=1b31a26007a745ea9a6b0cae9b94bbe4")
driver.maximize_window()

# change time if there are some error, which might 
# because loading time is too long and need more time until the picture show up
time_to_wait = 3

# check all library hits
all_library_hits_button = driver.find_element(By.XPATH, '/html/body/div/div[2]/table/tbody/tr[2]/td/table/tbody/tr[2]/td/div[4]/a[1]')
all_library_hits_button.click()
WebDriverWait(driver, time_to_wait).until(lambda s: s.find_element(By.ID, 'main.#Scan#_lowerinput').is_displayed()) 

for cluster_index in feature_list:
    cluster_index_lower = driver.find_element(By.ID, 'main.#Scan#_lowerinput')
    cluster_index_upper = driver.find_element(By.ID, 'main.#Scan#_upperinput')
    cluster_index_lower.send_keys(str(cluster_index))
    cluster_index_upper.send_keys(str(cluster_index))
    cluster_index_upper.send_keys(Keys.ENTER)
    try:
        view_mirror_buttom = driver.find_element(By.XPATH, '/html/body/div[3]/div[1]/table/tbody/tr[1]/td[1]/button[1]')
        WebDriverWait(driver, time_to_wait).until(lambda s: s.find_element(By.XPATH, '/html/body/div[3]/div[1]/table/tbody/tr[1]/td[1]/button[1]').is_displayed()) 
        view_mirror_buttom.click()
        WebDriverWait(driver, time_to_wait).until(lambda s: s.find_element(By.XPATH, '/html/body/div[3]/div[1]/table/tbody/tr[3]/td[2]/div/div/table/tbody/tr[3]/td/div[1]/canvas[2]').is_displayed())        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        driver.find_element(By.XPATH,'/html/body/div[3]/div[1]/table/tbody/tr[3]/td[2]/div/div/table/tbody/tr[3]').screenshot('all-library-hit-' + str(cluster_index) + '.png')
        print("mirror plot exist " + str(cluster_index))
        cluster_index_lower.clear()
        cluster_index_upper.clear()
    except NoSuchElementException:
        print('mirror plot not exist ' + str(cluster_index))
        cluster_index_lower.clear()
        cluster_index_upper.clear()

back_status_page_button = driver.find_element(By.XPATH, '/html/body/div[1]/div/table/tbody/tr[1]/td[2]/a')
back_status_page_button.click()

# check analog library hits
analog_library_hits_button = driver.find_element(By.XPATH, '/html/body/div/div[2]/table/tbody/tr[2]/td/table/tbody/tr[2]/td/div[4]/a[3]')
analog_library_hits_button.click()
WebDriverWait(driver, time_to_wait).until(lambda s: s.find_element(By.ID, 'main.#Scan#_lowerinput').is_displayed()) 

for cluster_index in feature_list:
    cluster_index_lower = driver.find_element(By.ID, 'main.#Scan#_lowerinput')
    cluster_index_upper = driver.find_element(By.ID, 'main.#Scan#_upperinput')
    cluster_index_lower.send_keys(str(cluster_index))
    cluster_index_upper.send_keys(str(cluster_index))
    cluster_index_upper.send_keys(Keys.ENTER)
    try:
        view_mirror_buttom = driver.find_element(By.XPATH, '/html/body/div[3]/div[1]/table/tbody/tr[1]/td[1]/img')
        WebDriverWait(driver, time_to_wait).until(lambda s: s.find_element(By.XPATH, '/html/body/div[3]/div[1]/table/tbody/tr[1]/td[1]/img').is_displayed()) 
        view_mirror_buttom.click()
        # some analog mirror plots take time to load
        time.sleep(time_to_wait)
        WebDriverWait(driver, time_to_wait).until(lambda s: s.find_element(By.XPATH, '/html/body/div[3]/div[1]/table/tbody/tr[2]/td[2]/div/div/table/tbody/tr[3]').is_displayed())        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        driver.find_element(By.XPATH,'/html/body/div[3]/div[1]/table/tbody/tr[2]/td[2]/div/div/table/tbody/tr[3]').screenshot('analog-library-hit-' + str(cluster_index) + '.png')
        print("analog mirror plot exist " + str(cluster_index))
        cluster_index_lower.clear()
        cluster_index_upper.clear()
    except NoSuchElementException:
        print('analog mirror plot not exist ' + str(cluster_index))
        cluster_index_lower.clear()
        cluster_index_upper.clear()

# close the browser after it finished
WebDriverWait(driver, time_to_wait)
driver.close()

```
