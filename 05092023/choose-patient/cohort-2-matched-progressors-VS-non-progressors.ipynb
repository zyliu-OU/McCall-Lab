```python
import numpy as np
import pandas as pd

# load the data
data_progress = pd.read_csv("progressors_data_fix_remove_unknown.csv")
data_progress_index_list = data_progress.index.tolist()

data_not_progress = pd.read_csv("non_progressors_data_fix_remove_unknown.csv")
data_not_progress_index_list = data_not_progress.index.tolist()
```

```python
# check how many IDs in each data.
print(data_progress_index_list)
print(data_not_progress_index_list)
```

```python
# create new empty lists
progress_id_matched_list = []
not_progress_id_matched_list = []

# one progressed patient correspond multiple not progressed patient, so use dictionary is perfect.
# create the dictionary to store the keys and values
dict_progress_key_notprogress_value = {}

# for loop in for loop, for each patient ID's age and sex from progressed list,
# see whether it matched with patient ID's age and sex from not progressed list.
# in this for loop, the "if" have two conditions: age difference smaller than 5 years, sex is matched,
# if the first stage of patient is same, then its matched. Because the each patient's stage is ordered by ABCD.
for progress_id in data_progress_index_list:
    # clean the not_progress_id_matched_list list so each loop will only store the not progressed IDs for that one specific progressed ID.
    not_progress_id_matched_list = []
    for not_progress_id in data_not_progress_index_list:
        if abs(data_progress.iloc[progress_id]['age'] - data_not_progress.iloc[not_progress_id]['age']) <= 5 and data_progress.iloc[progress_id]['sex'] == data_not_progress.iloc[not_progress_id]['sex'] and data_progress.iloc[progress_id]['stage'][2] == data_not_progress.iloc[not_progress_id]['stage'][2]:
            print(str(data_progress.iloc[progress_id]['patient_id']) + " and " + str(data_not_progress.iloc[not_progress_id]['patient_id']) + ' is matched')
            # store the multiple not progressed IDs into a list.
            not_progress_id_matched_list.append(data_not_progress.iloc[not_progress_id]['patient_id'])
    # store the list into that corresponding progressed ID.        
    dict_progress_key_notprogress_value[data_progress.iloc[progress_id]['patient_id']] = not_progress_id_matched_list
```

```python
# get the keys and values and tranform into lists
dict_keys_list = list(dict_progress_key_notprogress_value.keys())
dict_values_list = list(dict_progress_key_notprogress_value.values())
```

```python
# zip the lists and output the matched dataframe
matched_df = pd.DataFrame(list(zip(dict_keys_list, dict_values_list)), columns=['progress_id', 'not_progress_id'])
matched_df.to_csv('matched-data-non-progressors-single-row.csv',index=False)
```
