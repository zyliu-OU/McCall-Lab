```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read data and subset data.
dunn_data = pandas.read_csv('time-RVB-distance.csv')
group1C_to_other = dunn_data[dunn_data["name"].isin(["naive_VS_infected_50DPI", "naive_VS_infected_75DPI", "naive_VS_infected_142DPI"])]

plot_1 = sns.boxplot(y=group1C_to_other['name'] , x=group1C_to_other['Distance'], showfliers = False)
plot_1 = sns.stripplot(y=group1C_to_other['name'] , x=group1C_to_other['Distance'], color = "black", size=2, edgecolor="black", linewidth=.9)

plot_1.set_title("RVB")
plot_1.set_xlabel('Bray-Curtis distance')
plot_1.set_ylabel('')
plot_1.set_xlim([0, 1])
plt.savefig("distance-analysis-RVB.png", dpi=300, bbox_inches='tight')
```
