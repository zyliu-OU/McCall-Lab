```r
features_pos <- read.csv("treatment-LVB-metadata-feature-table.csv")
head(features_pos)
dim(features_pos)
```

```r
noMeta<-features_pos[,8:210]
features_wMeta<-cbind(features_pos[,1:7], noMeta)
head(noMeta)
dim(noMeta)

head(features_wMeta)
dim(features_wMeta)
```

```r
avg<- apply(features_wMeta[,8:ncol(features_wMeta)], 2, function(x) tapply(x, features_wMeta$Treatmentgroup, function(x) mean(x)))
avg
```

```r
BNZ_142DPI_mean <- avg[1,]
naive_142DPI_mean <- avg[2,]
infected_142DPI_mean <- avg[3,]
combo_142DPI_mean <- avg[4,]
```

```r
df <- data.frame(naive_142DPI_mean, infected_142DPI_mean, combo_142DPI_mean, BNZ_142DPI_mean)
write.csv(df, "LVB-mean-1C-3E-9-15.csv")
```
