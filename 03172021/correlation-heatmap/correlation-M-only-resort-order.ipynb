```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data_a = pd.read_csv("142DPI-M-reorder.csv")
data_a = data_a.set_index('MouseID')
sns.set(rc = {'figure.figsize':(35,25)})

g = sns.heatmap(data_a.corr(method='spearman'), vmin=-1, vmax=1, center= 0, cmap = 'coolwarm')

plt.savefig("M-correlation-heatmap-reordered.pdf", dpi = 300, bbox_inches = 'tight')
plt.savefig("M-correlation-heatmap-reordered.png", dpi = 300, bbox_inches = 'tight')
```
