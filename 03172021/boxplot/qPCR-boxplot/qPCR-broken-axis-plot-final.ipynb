```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import median

sns.set_theme(style="white")

dunn_data = pandas.read_csv('qPCR_data_fix_group_name.csv')
treatment_9_15_series = dunn_data[dunn_data["group"].isin(["naive", "infected", "combo", "BNZ"])]

f, (ax5, ax4, ax3, ax2, ax1) = plt.subplots(ncols=1, nrows=5, sharex=True)

sns.stripplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'], hue = treatment_9_15_series['position'], ax=ax1)
sns.stripplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'], hue = treatment_9_15_series['position'], ax=ax2)
sns.stripplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'], hue = treatment_9_15_series['position'], ax=ax3)
sns.stripplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'], hue = treatment_9_15_series['position'], ax=ax4)
sns.stripplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'], hue = treatment_9_15_series['position'], ax=ax5)

sns.pointplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'],estimator=median, color="black", join=False, ci=0, capsize=.7, scale=0, ax = ax1)
sns.pointplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'],estimator=median, color="black", join=False, ci=0, capsize=.7, scale=0, ax = ax2)
sns.pointplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'],estimator=median, color="black", join=False, ci=0, capsize=.7, scale=0, ax = ax3)
sns.pointplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'],estimator=median, color="black", join=False, ci=0, capsize=.7, scale=0, ax = ax4)
sns.pointplot(y = treatment_9_15_series['burden'] , x = treatment_9_15_series['group'],estimator=median, color="black", join=False, ci=0, capsize=.7, scale=0, ax = ax5)



ax5.set_ylim(60000, 150000)  
ax4.set_ylim(40000, 60000)  
ax3.set_ylim(7000, 40000)  
ax2.set_ylim(335, 7000)  
ax1.set_ylim(-100,335)



sns.set(rc={'figure.figsize':(5,10)})

sns.despine(ax=ax5, top = False, bottom=True, right = False)
sns.despine(ax=ax4, bottom=True, right = False)
sns.despine(ax=ax3, bottom=True, right = False)
sns.despine(ax=ax2, bottom=True, right = False)
sns.despine(ax=ax1, right = False)

ax1.yaxis.tick_left()
ax2.yaxis.tick_left()
ax3.yaxis.tick_left()
ax4.yaxis.tick_left()
ax5.yaxis.tick_left()

d = 2.  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot([1,1], transform=ax1.transAxes, **kwargs)

ax2.plot([0,0], transform=ax2.transAxes, **kwargs)
ax2.plot([1,1], transform=ax2.transAxes, **kwargs)

ax3.plot([0,0], transform=ax3.transAxes, **kwargs)
ax3.plot([1,1], transform=ax3.transAxes, **kwargs)

ax4.plot([0,0], transform=ax4.transAxes, **kwargs)
ax4.plot([1,1], transform=ax4.transAxes, **kwargs)

ax5.plot([0,0], transform=ax5.transAxes, **kwargs)

ax1.set_xlabel('')
ax1.set_ylabel('')

ax2.set_xlabel('')
ax2.set_ylabel('')

ax3.set_xlabel('')
ax3.set_ylabel('')

ax3.set_xlabel('')
ax3.set_ylabel('')

ax4.set_xlabel('')
ax4.set_ylabel('')

ax5.set_xlabel('')
ax5.set_ylabel('')


#remove extra legend and move one legend to right.


ax1.legend_.remove()
ax2.legend_.remove()
ax3.legend_.remove()
ax4.legend_.remove()
ax5.legend(bbox_to_anchor=(1.3,1), facecolor='white')
plt.savefig("qPCR-stripplot-broken-axis-final.pdf", dpi=300, bbox_inches='tight')
plt.savefig("qPCR-stripplot-broken-axis-final.png", dpi=300, bbox_inches='tight')
```
