```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read data and subset data.
dunn_data = pandas.read_csv('RVB-distance-data-for-dunn-test.csv')
group1C_to_other = dunn_data[dunn_data["group_name"].isin(["naive_VS_infected","naive_VS_combo", "naive_VS_BNZ"])]
group3E_to_other = dunn_data[dunn_data["group_name"].isin(["infected_VS_combo", "infected_VS_BNZ"])]

plt.style.use('default')
plt.rc('xtick',labelsize=12)
plt.rc('ytick',labelsize=20)
plot_1 = sns.boxplot(y=group1C_to_other['group_name'], x=group1C_to_other['Distance'], showfliers = False)
plot_1 = sns.stripplot(y=group1C_to_other['group_name'], x=group1C_to_other['Distance'], color = "black", size=2, edgecolor="black", linewidth=.9)

plot_1.patches[0].set_facecolor('#e1812c')
plot_1.patches[1].set_facecolor('#3a923a')
plot_1.patches[2].set_facecolor('#c03d3e')

plot_1.set_title("RVB naive VS other treatments",fontsize=20)
plot_1.set_ylabel('')
plot_1.set_xlabel('Bray-Curtis distance', fontsize=20)
plot_1.set_xlim([0, 0.8])
plt.savefig("distance_analysis_1C_RVB.png", dpi=300, bbox_inches='tight')
```

```python
plt.style.use('seaborn-bright')
plot_2 = sns.boxplot(y=group3E_to_other['group_name'], x=group3E_to_other['Distance'], showfliers = False)
plot_2 = sns.stripplot(y=group3E_to_other['group_name'], x=group3E_to_other['Distance'], color = "black", size=2, edgecolor="black", linewidth=.9)

plot_2.set_title("RVB infected VS other treatments",fontsize=20)
plot_2.set_ylabel('')
plot_2.set_xlabel('Bray-Curtis distance', fontsize=20)
plot_2.set_xlim([0, 0.8])
plt.savefig("distance_analysis_3E_RVB.png", dpi=300, bbox_inches='tight')
```
