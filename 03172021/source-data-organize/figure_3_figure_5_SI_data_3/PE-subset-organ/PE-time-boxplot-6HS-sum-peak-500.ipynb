```python

import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data_a = pandas.read_csv('03172021-PE-M-F-time-sum-peak.csv')

data_LA = data_a[data_a["organ"].isin(["LA"])]
data_RA = data_a[data_a["organ"].isin(["RA"])]
data_LVT = data_a[data_a["organ"].isin(["LV_top"])]
data_LVB = data_a[data_a["organ"].isin(["LV_bottom"])]
data_RVT = data_a[data_a["organ"].isin(["RV_top"])]
data_RVB = data_a[data_a["organ"].isin(["RV_bottom"])]

mass_list = ['mass_500']
print(mass_list)

# change the theme of the plot
sns.set_theme(style="white")

plt.rc('xtick',labelsize=16)
plt.rc('ytick',labelsize=16)

mass_list_number = 1
for mass in mass_list:
    fig, axes = plt.subplots(2, 3, figsize=(18, 10), constrained_layout=True)
    fig.suptitle(mass)
    g1 = sns.boxplot(ax=axes[0, 0], y=data_LA[str(mass)], x=data_LA['DPI'],hue=data_LA['status'], palette="tab10", showfliers = False)
    g1 = sns.stripplot(ax=axes[0, 0], y=data_LA[str(mass)], x=data_LA['DPI'],hue=data_LA['status'], dodge=True, color = "black", size=2, edgecolor="black", linewidth=.9)
    axes[0, 0].set_title("LA", fontsize=16)
    axes[0, 0].set_xlabel('DPI', fontsize=16)
    axes[0, 0].set_ylabel('Normalized peak area', fontsize=16)
    g2 = sns.boxplot(ax=axes[0, 1], y=data_RA[str(mass)], x=data_RA['DPI'],hue=data_RA['status'], palette="tab10", showfliers = False)
    g2 = sns.stripplot(ax=axes[0, 1], y=data_RA[str(mass)], x=data_RA['DPI'],hue=data_RA['status'], dodge=True, color = "black", size=2, edgecolor="black", linewidth=.9)
    axes[0, 1].set_title("RA", fontsize=16)
    axes[0, 1].set_xlabel('DPI', fontsize=16)
    axes[0, 1].set_ylabel('Normalized peak area', fontsize=16)
    axes[0, 1].set_ylim([-0.0001, 0.03])
    g3 = sns.boxplot(ax=axes[0, 2], y=data_LVT[str(mass)], x=data_LVT['DPI'],hue=data_LVT['status'], palette="tab10", showfliers = False)
    g3 = sns.stripplot(ax=axes[0, 2], y=data_LVT[str(mass)], x=data_LVT['DPI'],hue=data_LVT['status'], dodge=True, color = "black", size=2, edgecolor="black", linewidth=.9)
    axes[0, 2].set_title("LVT", fontsize=16)
    axes[0, 2].set_xlabel('DPI', fontsize=16)
    axes[0, 2].set_ylabel('Normalized peak area', fontsize=16)
    g4 = sns.boxplot(ax=axes[1, 0], y=data_LVB[str(mass)], x=data_LVB['DPI'],hue=data_LVB['status'], palette="tab10", showfliers = False)
    g4 = sns.stripplot(ax=axes[1, 0], y=data_LVB[str(mass)], x=data_LVB['DPI'],hue=data_LVB['status'], dodge=True, color = "black", size=2, edgecolor="black", linewidth=.9)
    axes[1, 0].set_title("LVB", fontsize=16)
    axes[1, 0].set_xlabel('DPI', fontsize=16)
    axes[1, 0].set_ylabel('Normalized peak area', fontsize=16)
    axes[1, 0].set_ylim([-0.0001, 0.014])
    g5 = sns.boxplot(ax=axes[1, 1], y=data_RVT[str(mass)], x=data_RVT['DPI'],hue=data_RVT['status'], palette="tab10", showfliers = False)
    g5 = sns.stripplot(ax=axes[1, 1], y=data_RVT[str(mass)], x=data_RVT['DPI'],hue=data_RVT['status'], dodge=True, color = "black", size=2, edgecolor="black", linewidth=.9)
    axes[1, 1].set_title("RVT", fontsize=16)
    axes[1, 1].set_xlabel('DPI', fontsize=16)
    axes[1, 1].set_ylabel('Normalized peak area', fontsize=16)
    g6 = sns.boxplot(ax=axes[1, 2], y=data_RVB[str(mass)], x=data_RVB['DPI'],hue=data_RVB['status'], palette="tab10", showfliers = False)
    g6 = sns.stripplot(ax=axes[1, 2], y=data_RVB[str(mass)], x=data_RVB['DPI'],hue=data_RVB['status'], dodge=True, color = "black", size=2, edgecolor="black", linewidth=.9)
    axes[1, 2].set_title("RVB", fontsize=16)
    axes[1, 2].set_xlabel('DPI', fontsize=16)
    axes[1, 2].set_ylabel('Normalized peak area', fontsize=16)
    axes[1, 2].set_ylim([-0.0001, 0.016])
    sns.move_legend(g1, "upper right")
    sns.move_legend(g2, "upper right")
    sns.move_legend(g3, "upper right")
    sns.move_legend(g4, "upper right")
    sns.move_legend(g5, "upper right")
    sns.move_legend(g6, "upper right")
    g1.legend([],[], frameon=False)
    g2.legend([],[], frameon=False)
    g3.legend([],[], frameon=False)
    g4.legend([],[], frameon=False)
    g5.legend([],[], frameon=False)
    g6.legend([],[], frameon=False)
    plt.savefig("6-HS-boxplot-PE-time-sum-500-" + str(mass) + ".png", dpi=300, bbox_inches='tight')
    mass_list_number += 1
```
