```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="white")

# read data and subset data.
data_1 = pandas.read_csv('time-pseudo-F.csv')
plot_1 = sns.lineplot(y=data_1['pseudo_F'] , x=data_1['DPI'], hue=data_1['heart section'], palette="tab10")
plot_1.set_title("")
plot_1.set_xlabel('DPI')
plot_1.set_ylabel('pseudo-F')
x_ticks = [50,75,142]
plt.xticks(ticks=x_ticks)
sns.move_legend(plot_1, "upper right", bbox_to_anchor=(1.28,1.02))
plt.savefig("lineplot.png", dpi=300, bbox_inches='tight')
plt.savefig("lineplot.pdf", dpi=300, bbox_inches='tight')
```
