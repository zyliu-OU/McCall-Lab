```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read data and subset data.
dunn_data = pandas.read_csv('treatment-LVB-distance.csv')
group1C_to_other = dunn_data[dunn_data["group_name"].isin(["naive_VS_naive", "naive_VS_infected", "naive_VS_combo", "naive_VS_BNZ"])]
group3E_to_other = dunn_data[dunn_data["group_name"].isin(["infected_VS_infected", "infected_VS_combo", "infected_VS_BNZ"])]

plt.style.use('default')
plot_1 = sns.boxplot(y=group1C_to_other['group_name'] , x=group1C_to_other['Distance'], showfliers = False)
plot_1 = sns.stripplot(y=group1C_to_other['group_name'] , x=group1C_to_other['Distance'], color = "black", size=2, edgecolor="black", linewidth=.9)

plot_1.set_title("LVB naive VS other treatments")
plot_1.set_xlabel('Bray-Curtis distance')
plot_1.set_ylabel('')
plot_1.set_xlim([0, 1])
plt.savefig("treatment_1C_LVB.png", dpi=300, bbox_inches='tight')
```

```python
plt.style.use('seaborn-bright')
plot_2 = sns.boxplot(y=group3E_to_other['group_name'] , x=group3E_to_other['Distance'], showfliers = False)
plot_2 = sns.stripplot(y=group3E_to_other['group_name'] , x=group3E_to_other['Distance'], color = "black", size=2, edgecolor="black", linewidth=.9)

plot_2.set_title("LVB infected VS other treatments")
plot_2.set_xlabel('Bray-Curtis distance')
plot_2.set_ylabel('')
plot_2.set_xlim([0, 1])
plt.savefig("treatment_3E_LVB.png", dpi=300, bbox_inches='tight')
```
