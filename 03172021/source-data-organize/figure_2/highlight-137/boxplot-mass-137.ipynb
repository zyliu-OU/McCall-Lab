```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data_LA = pandas.read_csv('LA_selected_mass_fix.csv')
data_RA = pandas.read_csv('RA_selected_mass_fix.csv')
data_LVT = pandas.read_csv('LVT_selected_mass_fix.csv')
data_LVB = pandas.read_csv('LVB_selected_mass_fix.csv')
data_RVT = pandas.read_csv('RVT_selected_mass_fix.csv')
data_RVB = pandas.read_csv('RVB_selected_mass_fix.csv')

plt.rc('xtick',labelsize=16)
plt.rc('ytick',labelsize=16)

fig, axes = plt.subplots(2, 3, figsize=(18, 10), constrained_layout=True)
fig.suptitle("m/z 137.046 RT 0.35 min annotated as hypoxanthine")
g1 = sns.boxplot(ax=axes[0, 0], y=data_LA['mass_137'], x=data_LA['DPI'],hue=data_LA['status'], showfliers = False)
g1 = sns.stripplot(ax=axes[0, 0], y=data_LA['mass_137'], x=data_LA['DPI'],hue=data_LA['status'], color = "black", size=2, dodge=True, edgecolor="black", linewidth=.9)
axes[0, 0].set_title("LA", fontsize=16)
axes[0, 0].set_xlabel('DPI', fontsize=16)
axes[0, 0].set_ylabel('Normalized peak area', fontsize=16)

g2 = sns.boxplot(ax=axes[0, 1], y=data_RA['mass_137'], x=data_RA['DPI'],hue=data_RA['status'], showfliers = False)
g2 = sns.stripplot(ax=axes[0, 1], y=data_RA['mass_137'], x=data_RA['DPI'],hue=data_RA['status'], color = "black", size=2, dodge=True, edgecolor="black", linewidth=.9)
axes[0, 1].set_title("RA", fontsize=16)
axes[0, 1].set_xlabel('DPI', fontsize=16)
axes[0, 1].set_ylabel('Normalized peak area', fontsize=16)

g3 = sns.boxplot(ax=axes[0, 2], y=data_LVT['mass_137'], x=data_LVT['DPI'],hue=data_LVT['status'], showfliers = False)
g3 = sns.stripplot(ax=axes[0, 2], y=data_LVT['mass_137'], x=data_LVT['DPI'],hue=data_LVT['status'], color = "black", size=2, dodge=True, edgecolor="black", linewidth=.9)
axes[0, 2].set_title("LVT", fontsize=16)
axes[0, 2].set_xlabel('DPI', fontsize=16)
axes[0, 2].set_ylabel('Normalized peak area', fontsize=16)

g4 = sns.boxplot(ax=axes[1, 0], y=data_LVB['mass_137'], x=data_LVB['DPI'],hue=data_LVB['status'], showfliers = False)
g4 = sns.stripplot(ax=axes[1, 0], y=data_LVB['mass_137'], x=data_LVB['DPI'],hue=data_LVB['status'], color = "black", size=2, dodge=True, edgecolor="black", linewidth=.9)
axes[1, 0].set_title("LVB", fontsize=16)
axes[1, 0].set_xlabel('DPI', fontsize=16)
axes[1, 0].set_ylabel('Normalized peak area', fontsize=16)

g5 = sns.boxplot(ax=axes[1, 1], y=data_RVT['mass_137'], x=data_RVT['DPI'],hue=data_RVT['status'], showfliers = False)
g5 = sns.stripplot(ax=axes[1, 1], y=data_RVT['mass_137'], x=data_RVT['DPI'],hue=data_RVT['status'], color = "black", size=2, dodge=True, edgecolor="black", linewidth=.9)
axes[1, 1].set_title("RVT", fontsize=16)
axes[1, 1].set_xlabel('DPI', fontsize=16)
axes[1, 1].set_ylabel('Normalized peak area', fontsize=16)

g6 = sns.boxplot(ax=axes[1, 2], y=data_RVB['mass_137'], x=data_RVB['DPI'],hue=data_RVB['status'], showfliers = False)
g6 = sns.stripplot(ax=axes[1, 2], y=data_RVB['mass_137'], x=data_RVB['DPI'],hue=data_RVB['status'], color = "black", size=2, dodge=True, edgecolor="black", linewidth=.9)
axes[1, 2].set_title("RVB", fontsize=16)
axes[1, 2].set_xlabel('DPI', fontsize=16)
axes[1, 2].set_ylabel('Normalized peak area', fontsize=16)

sns.move_legend(g1, "upper right")
sns.move_legend(g2, "upper right")
sns.move_legend(g3, "upper right")
sns.move_legend(g4, "upper right")
sns.move_legend(g5, "upper right")
sns.move_legend(g6, "upper right")

g1.legend([],[], frameon=False)
g2.legend([],[], frameon=False)
g3.legend([],[], frameon=False)
g4.legend([],[], frameon=False)
g5.legend([],[], frameon=False)
g6.legend([],[], frameon=False)

plt.savefig("boxplot-mass-137.png", dpi=300, bbox_inches='tight')
```
